```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 5000

# Features
cloud_cover = np.random.uniform(0, 100, n)
sunlight_hours = np.random.uniform(0, 12, n)
solar_radiation = np.random.uniform(200, 1000, n)
panel_efficiency = np.random.uniform(0.15, 0.22, n)

# Synthetic realistic power formula
generated_power = (
    solar_radiation
    * sunlight_hours
    * panel_efficiency
    * (1 - cloud_cover / 120)
)

# Add moderate noise (important!)
noise = np.random.normal(0, 100, n)
generated_power = generated_power + noise

generated_power = np.maximum(generated_power, 0)

# Create dataframe
synthetic_data = pd.DataFrame({
    "cloud_cover": cloud_cover,
    "direct_sunlight_hours": sunlight_hours,
    "solar_radiation": solar_radiation,
    "panel_efficiency": panel_efficiency,
    "generated_power_kw": generated_power
})

# Save CSV
synthetic_data.to_csv("synthetic_solar_dataset.csv", index=False)

synthetic_data.head()
```
